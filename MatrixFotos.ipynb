```python
import os
import numpy as np
from keras.preprocessing import image
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
```

```python
testfoto = []

path='Fotos'
for(path,dirs,files) in os.walk(path):
    if path == 'Fotos/test/images': 
        for i in range(0,len(files)):
            chorro = 'Fotos/test/images/' + str(files[i])
            img_path = chorro
            img = image.load_img(img_path, target_size=(224, 224))
            testfoto.append(img)
            
```

```python
ltestfoto = []

for m in range(len(testfoto)):
    ltestfoto.append(np.array(testfoto[m]))
```

```python
plt.imshow(ltestfoto[0])
```

```python
trainfoto = []

path='Fotos'
for(path,dirs,files) in os.walk(path):
    if path == 'Fotos/train/images':
        for j in range(0,len(files)):
            chorro = 'Fotos/train/images/' + str(files[j])
            img_path = chorro
            img = image.load_img(img_path, target_size=(224, 224))
            trainfoto.append(img)
```

```python
ltrainfoto = []

for p in range(len(trainfoto)):
    ltrainfoto.append(np.array(trainfoto[p]))
```

```python
valfoto = []

path='Fotos'
for(path,dirs,files) in os.walk(path):
    if path == 'Fotos/val/images':
        for k in range(0,len(files)):
            chorro = 'Fotos/val/images/' + str(files[k])
            img_path = chorro
            img = image.load_img(img_path, target_size=(224, 224))
            valfoto.append(img)
```

```python
lvalfoto = []

for w in range(len(valfoto)):
    lvalfoto.append(np.array(valfoto[w]))
```

```python
import pickle

pickle_out_test = open("testfoto.pickle", "wb")
pickle.dump(ltestfoto, pickle_out_test)
pickle_out_test.close()
```

```python
pickle_out_train = open("trainfoto.pickle", "wb")
pickle.dump(ltrainfoto, pickle_out_train)
pickle_out_train.close()
```

```python
pickle_out_val = open("valfoto.pickle", "wb")
pickle.dump(lvalfoto, pickle_out_val)
pickle_out_val.close()
```
